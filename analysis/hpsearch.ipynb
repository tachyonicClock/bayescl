```python
import optuna
from common import PLOT_METHODS, DATASET_TO_LABEL, COLORS, METRIC_TO_LABEL, METHOD_TO_LABEL
from matplotlib import pyplot as plt
import os

optuna_storage = optuna.storages.RDBStorage(os.environ.get("OPTUNA_STORAGE"))
```

```python
LIMIT = 10
width = 3
fig, axies = plt.subplots(figsize=(width * 3, width), ncols=3)

for i, (dataset, label) in enumerate(DATASET_TO_LABEL.items()):
    ax = axies[i]
    for method in PLOT_METHODS:
        study_name = f"bayescl/hp/{dataset}/{method}"
        
        try:
            study = optuna.load_study(study_name=study_name, storage=optuna_storage)
            df = study.trials_dataframe()
            df = df[df.state == "COMPLETE"][:LIMIT]
            ax.scatter(
                df["values_0"],
                df["values_1"],
                color=COLORS.get(method, "red")
            )
            ax.set_title(label)

            # Make figure a square
            ax.set_box_aspect(1)

            print(study_name, len(df))
        except KeyError:
            print(study_name, 0)
            continue

    axies[1].set_xlabel(METRIC_TO_LABEL.get("accuracy_seen_avg"))
    axies[0].set_ylabel(METRIC_TO_LABEL.get("ece_seen_avg"))

# Add legend to the right of the figure
handles = [plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=COLORS.get(method, "red"), markersize=10) for method in PLOT_METHODS]
labels = [METHOD_TO_LABEL.get(method, method) for method in PLOT_METHODS]
fig.legend(handles, labels, loc='center right', frameon=False)
plt.tight_layout(rect=[0, 0, 0.86, 1])
```
