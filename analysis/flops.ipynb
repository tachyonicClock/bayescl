```python
# https://alessiodevoto.github.io/Compute-Flops-with-Pytorch-built-in-flops-counter/
from typing import Tuple, Union

import torch
from torch.utils.flop_counter import FlopCounterMode
from torchvision.models import resnet18


def get_GFLOP_count(model, inp: Union[torch.Tensor, Tuple], with_backward=False):

    istrain = model.training
    model.eval()

    inp = inp if isinstance(inp, torch.Tensor) else torch.randn(inp)

    flop_counter = FlopCounterMode()
    with flop_counter:
        if with_backward:
            model(inp).sum().backward()
        else:
            model(inp)
    total_flops = flop_counter.get_total_flops()
    if istrain:
        model.train()
    return total_flops / 2 / 1e9


model = resnet18()
shape = (1, 3, 224, 224)
gflop = get_GFLOP_count(model, shape, True)
print(f"GFLOP: {gflop:.2f} GFLOPs")
```
